## Multicollinearity using VIF
* COlumsn :'Age','EmpJobInvolvement' ,'EmpLastSalaryHikePercent','EmpWorkLifeBalance' shows Multicollinearity i.e VIF >10
* But dropping all above columns negatively affecting the accuracy hence dropping only 'Age' column

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)

calc_vif(X)
```

```python
# Drop the columns which shows Multicollinearity i.e VIF >10
del_col =['Age']
df = df.loc[: , ~df.columns.isin(del_col)]
df.head(2)
```

## Feature Engineering 

# Label Encoding

```python
df[Catcol_list].head()

for i in Catcol_list_toEncode:
    val=df[i].value_counts().count()
    print("{}  : {}".format(i,val))
    print("="*40)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder =LabelEncoder()
for i in Catcol_list_toEncode :
    df[i] = encoder.fit_transform(df[i])
    
df[Catcol_list_toEncode].head(2)
df['PerformanceRating'] = encoder.fit_transform(df['PerformanceRating'])
```

## Train Test Split

```python
X1 = df.loc[:,df.columns != 'PerformanceRating']
y1 = df['PerformanceRating']

from imblearn.combine import SMOTETomek
smk =SMOTETomek(random_state=42)
X, y = smk.fit_sample(X1,y1)

print(X.shape)
print(y.shape)
```

# PCA

```python
from sklearn.preprocessing import StandardScaler

pca = StandardScaler()
X_std = pca.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA()
X_pca =pca.fit_transform(X_std)

print(pca.explained_variance_)
print(pca.explained_variance_ratio_)
```

```python
li =pd.DataFrame(data=pca.explained_variance_ratio_ )
li=li*100
li[1] = df.iloc[:,0:-1].columns
li.columns = ['pca_values' ,'Col_names']
li
```

```python
df_skb =li.sort_values(by='pca_values',ascending=False)
X_pcb=list(df_skb.iloc[0:11,1])
X_pcb
```

```python
rcParams['figure.figsize']=10,5
groupedvalues=li.groupby('pca_values').sum().reset_index()
groupedvalues=groupedvalues.sort_values('pca_values').reset_index()

ax = sns.barplot(x =li['pca_values'] ,y=li['Col_names'],data=groupedvalues)

```

## Feature Selection by : SelectKBest

```python
from sklearn.feature_selection import SelectKBest ,chi2
from sklearn import preprocessing

X_norm = pd.DataFrame(preprocessing.normalize(X))

selectK = SelectKBest(score_func = chi2 , k=8)
selectK.fit(X_norm,y)

df_skb = pd.DataFrame({'feature' : X.columns , 'score' :selectK.scores_ })
df_skb.sort_values(by='score',ascending=True).plot(kind='barh',x='feature',y='score')
```

```python
# From the above graph we are select best 8 features for modelling

df_skb =df_skb.sort_values(by='score',ascending=False)
X_skb=list(df_skb.iloc[0:8,0])
X_skb
```

## Modelling

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score , classification_report ,confusion_matrix ,precision_recall_curve ,auc
from sklearn.model_selection import KFold ,cross_val_score
```

```python
#prepare models

models=[]

models.append(('KNN',KNeighborsClassifier()))
models.append(('CART',DecisionTreeClassifier()))
models.append(('SVM',SVC()))
models.append(('XGBoost',XGBClassifier()))
models.append(('RF',RandomForestClassifier()))

score='accuracy'
results=[]
names=[]

for n ,m in models:
    kfold = KFold(n_splits=10 , random_state=10)
    
    cv_results = cross_val_score(m ,X,y,scoring=score,cv=kfold ) # m :model (model ,X,y,scoring,cv)
    results.append(cv_results)
    names.append(n)
    
    msg ="%s : %f (%f) " % (n,cv_results.mean(),cv_results.std())
    print(msg)
```

```python
#boxplot algorithm comparision
rcParams['figure.figsize'] =15,5
fig=plt.figure()
fig.suptitle("Algorithm Comparision")
ax=fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
```

## KFold CrossValidation

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=12,shuffle=True, random_state=45) 
for train_index, test_index in kf.split(X[X_pcb],y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index] 
    y_train, y_test = y[train_index], y[test_index]


print(X_train.shape)
print(y_train.shape)
```

```python
X_train.head()
```

##  XGBClassifer based on PCA

```python
model_XGB_SKB = XGBClassifier(booster='gbtree',
                             learning_rate=0.1,
                             n_estimators=1000,
                             max_depth=12)
model_XGB_SKB.fit(X_train,y_train)
y_predict4 = pd.DataFrame(model_XGB_SKB.predict(X_test),columns=['Status'])

print("Training Accuracy",model_XGB_SKB.score(X_train,y_train))
print("Testing Accuracy",model_XGB_SKB.score(X_test,y_test))
print("Acuuracy of XGBClassifier with PCA featues : {} {}".format(accuracy_score(y_test,y_predict4).round(4)*100 , "%"))
print("-"*60)
print(classification_report(y_test,y_predict4))
```

## XGBClassifier based on SelectKBest

```python
model_XGB_SKB = XGBClassifier(booster='gbtree',
                             learning_rate=0.1,
                             n_estimators=1000,
                             max_depth=12,
                             random_state=31)
model_XGB_SKB.fit(X_train,y_train)
y_predict4 = pd.DataFrame(model_XGB_SKB.predict(X_test),columns=['Status'])

print("Training Accuracy",model_XGB_SKB.score(X_train,y_train))
print("Testing Accuracy",model_XGB_SKB.score(X_test,y_test))
print("-"*60)
print("Acuuracy of XGBClassifier with SelectKBest featues : {} {}".format(accuracy_score(y_test,y_predict4).round(4)*100 , "%"))
print("-"*60)
print(classification_report(y_test,y_predict4))

```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(max_depth= 15, n_estimators= 1000,min_samples_split=4)

model_rf.fit(X_train,y_train)
y_predict5 = pd.DataFrame(model_rf.predict(X_test),columns=['Status'])

print("Training Accuracy",model_rf.score(X_train,y_train))
print("Testing Accuracy",model_rf.score(X_test,y_test))
print("Acuuracy of RandommForest with SelectKBest featues : {} {}".format(accuracy_score(y_test,y_predict4).round(4)*100 , "%"))
print("-"*60)
print(classification_report(y_test,y_predict5))
```
