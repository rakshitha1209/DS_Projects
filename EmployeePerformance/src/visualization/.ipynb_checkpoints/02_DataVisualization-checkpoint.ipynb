## Listing columns into Continous n Categorical

* Catcol_list_toEncode : Categorical data need to be encoded
* Catcol_list  : list of all categorical data 
* Concol_list  : list of all continous data

```python
#Check for null values
df.isnull().sum()

#Listing categorical n continous columns
Catcol_list_toEncode = ['Gender' ,'EducationBackground' ,'MaritalStatus','EmpDepartment' ,
                        'BusinessTravelFrequency','OverTime' ,'Attrition' ]
Catcol_list = ['Gender' ,'EducationBackground' ,'MaritalStatus','EmpDepartment' ,
               'BusinessTravelFrequency','OverTime' ,'Attrition','EmpEducationLevel' ,'EmpEnvironmentSatisfaction' ,
               'EmpJobInvolvement','EmpJobLevel','EmpJobSatisfaction','NumCompaniesWorked',
               'EmpRelationshipSatisfaction','TrainingTimesLastYear' ,'EmpWorkLifeBalance' ,'PerformanceRating']
Concol_list = ['Age', 'DistanceFromHome', 'EmpHourlyRate', 'EmpLastSalaryHikePercent',
       'TotalWorkExperienceInYears', 'ExperienceYearsAtThisCompany',
       'ExperienceYearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']
```

# EDA

## Univariant Analysis of Categorical variables

```python
fig, ax = plt.subplots(3, 6,figsize=(20, 10))
for variable, subplot in zip(Catcol_list, ax.flatten()):
    sns.countplot(df[variable], ax=subplot)
    for label in subplot.get_xticklabels():
        label.set_rotation(90)
```

# Univariate analysis of Continous variables

* Data distribution shows that data is not into gaussian distribution

```python
fig, ax = plt.subplots(3, 3, figsize=(20, 10))
for variable, subplot in zip(Concol_list, ax.flatten()):
    sns.countplot(df[variable], ax=subplot)
    for label in subplot.get_xticklabels():
        label.set_rotation(90)
```

# Bivariant Analysis

#### PerformanceRating Vs Age
###### We can observe that 
* majority of the people belong to mid-age( 27-38) shows better performance
* Employees who are in begining of career and end of career shows resonably less performance
###### Insights :
* It is advised to train the young generation & motivate them to improve their performance.
* Employees near to the end of their career usually in higher positions else few will  failed to reach higher positions due t some constraints hence no spirit left in elavating the performance.

```python
rcParams['figure.figsize'] =20,5
bv_plot = pd.crosstab(df['Age'] , df['PerformanceRating'])
bv_plot.plot(kind='bar' ,title='PerformanceRating Vs Age')
```
